```python
#1
# Generate data
x = np.linspace(-3, 3, 100)
y = x**2

# Create line plot
plt.plot(x, y, linestyle='dashed', color='blue', label=r'$y = x^2$')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadratic Function Plot')
plt.legend()

# Display the plot
plt.show()
```

```python
#2
# Product sales data
products = ['A', 'B', 'C']
sales = [120, 150, 90]

# Create bar chart
plt.bar(products, sales, color=['red', 'green', 'blue'], label='Sales')

# Add labels and title
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Product Sales Bar Chart')
plt.legend()

# Display the plot
plt.show()
```

```python
#3
# Generate random data
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)

# Create scatter plot with different colors
plt.scatter(x, y, c=np.where(y >= 0, 'blue', 'red'), label='Random Data')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Positive/Negative Coloring')
plt.legend()

# Display the plot
plt.show()
```

```python
#4
# Generate random data
data = np.random.randn(500)

# Create histogram with custom bin edges
plt.hist(data, bins=np.arange(-3, 3, 0.5), color='green', alpha=0.7, edgecolor='black')

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Custom Binned Histogram')
plt.grid(True)
```

```python
#5
# Expense distribution
expenses = ['Rent', 'Food', 'Transportation']
amounts = [40, 30, 20]

# Create pie chart
plt.pie(amounts, labels=expenses, autopct='%1.1f%%', colors=['orange', 'green', 'blue'])

# Add title
plt.title('Expense Distribution')
```

```python
# Sales distribution for four quarters
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
sales = [30, 25, 20, 25]
explode = (0, 0, 0.1, 0)  # Explode the third quarter

# Create pie chart with customization
plt.pie(sales, labels=quarters, autopct='%1.1f%%', colors=['blue', 'green', 'orange', 'red'], explode=explode)

# Add title
plt.title('Sales Distribution by Quarters')
```

```python
#7
# Generate data
x = np.linspace(-2, 2, 100)
y_exp_neg = np.exp(-x)
y_exp_pos = np.exp(x)

# Create line plot with legend and different line styles
plt.plot(x, y_exp_neg, label=r'$e^{-x}$', linestyle='dashed', color='green')
plt.plot(x, y_exp_pos, label=r'$e^{x}$', linestyle='dashdot', color='purple')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Exponential Functions')
plt.legend()
```

```python
# City population data
cities = ['A', 'B', 'C']
population = [5000000, 7000000, 3000000]

# Create horizontal bar chart
plt.barh(cities, population, color=['blue', 'green', 'orange'], label='Population')

# Add labels and title
plt.xlabel('Population')
plt.ylabel('Cities')
plt.title('City Population Distribution')
plt.legend()
```

```python
#9
# Generate data
x = np.linspace(-5, 5, 100)
y_linear = x
y_absolute = np.abs(x)

# Create line plot with legend and different line colors
plt.plot(x, y_linear, label=r'$y = x$', color='purple')
plt.plot(x, y_absolute, label=r'$y = |x|$', color='orange')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Linear and Absolute Value Functions')
plt.legend()
```

```python
#10
# Scatter plot data
x1 = [1, 2, 3, 4]
y1 = [2, 4, 3, 1]

x2 = [1, 2, 3, 4]
y2 = [4, 2, 3, 1]

# Create scatter plot with different markers and colors
plt.scatter(x1, y1, marker='^', color='red', label='Red Triangles')
plt.scatter(x2, y2, marker='o', color='blue', label='Blue Circles')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Different Markers and Colors')
plt.legend()
```

```python
#11
# Exam score data
class_a = np.random.normal(70, 10, 100)
class_b = np.random.normal(75, 8, 100)
class_c = np.random.normal(80, 12, 100)

# Create boxplot
plt.boxplot([class_a, class_b, class_c], labels=['Class A', 'Class B', 'Class C'], patch_artist=True)

# Customize colors
colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(plt.boxplot([class_a, class_b, class_c], patch_artist=True)['boxes'], colors):
    patch.set_facecolor(color)

# Add labels and title
plt.xlabel('Classes')
plt.ylabel('Exam Scores')
plt.title('Boxplot of Exam Scores by Class')
```

```python
#12
# Monthly expense data for two individuals
categories = ['Rent', 'Utilities', 'Groceries']
expenses_person1 = [1200, 200, 300]
expenses_person2 = [1000, 250, 350]

# Create stacked bar chart
bar_width = 0.5
plt.bar(categories, expenses_person1, color='lightblue', label='Person 1', width=bar_width)
plt.bar(categories, expenses_person2, bottom=expenses_person1, color='lightgreen', label='Person 2', width=bar_width)

# Add labels and title
plt.xlabel('Expense Categories')
plt.ylabel('Monthly Expenses')
plt.title('Stacked Bar Chart of Monthly Expenses')
plt.legend()
```

```python
#13
# Test score data
group_1 = np.random.normal(75, 10, 100)
group_2 = np.random.normal(70, 8, 100)

# Create horizontal boxplot
plt.boxplot([group_1, group_2], labels=['Group 1', 'Group 2'], vert=False, patch_artist=True)

# Customize colors
colors = ['lightpink', 'lightyellow']
for patch, color in zip(plt.boxplot([group_1, group_2], vert=False, patch_artist=True)['boxes'], colors):
    patch.set_facecolor(color)

# Add title
plt.title('Horizontal Boxplot of Test Scores by Group')
```

```python
#14
# Daily activity data for three weeks
activities = ['Work', 'Exercise', 'Leisure']
week1 = [8, 1, 5]
week2 = [7, 2, 6]
week3 = [9, 1.5, 4]

# Create stacked bar chart
bar_width = 0.5
plt.bar(activities, week1, color='lightblue', label='Week 1', width=bar_width)
plt.bar(activities, week2, bottom=week1, color='lightgreen', label='Week 2', width=bar_width)
plt.bar(activities, week3, bottom=np.array(week1)+np.array(week2), color='lightcoral', label='Week 3', width=bar_width)

# Add labels and title
plt.xlabel('Activities')
plt.ylabel('Hours')
plt.title('Stacked Bar Chart of Daily Activities')
plt.legend()
```
