```python
%matplotlib inline
import torch
from d2l import torch as d2l
```

 

Note that this is still linear since it's an affine function:



What we also need is a non-linear activation function such as ReLU $\sigma(x) = max(0,x) $ the output of activator functions are called *activations*

```python
x = torch.arange(-8.0, 8.0, 0.1, requires_grad=True)
y = torch.relu(x)
d2l.plot(x.detach(), y.detach(), 'x', 'relu(x)', figsize=(5, 2.5))
```

```python
y.backward(torch.ones_like(x), retain_graph=True)
d2l.plot(x.detach(), x.grad, 'x', 'grad of relu', figsize=(5, 2.5))
```
